```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Generate synthetic data
np.random.seed(0)
data = np.random.binomial(n=1, p=0.3, size=100)  # Binary data with success probability 0.3

# Define the likelihood function (Bernoulli distribution)
def likelihood(theta, data):
    return np.prod(theta**data * (1-theta)**(1-data))

# Define parameters for the Beta prior
alpha = 2  # Alpha parameter for the Beta distribution
beta_ = 2  # Beta parameter for the Beta distribution

# Generate a range of theta values
theta_values = np.linspace(0, 1, 1000)

# Calculate the Beta prior distribution
prior = beta.pdf(theta_values, alpha, beta_)

# Calculate the likelihood
likelihoods = [likelihood(theta, data) for theta in theta_values]

# Calculate the unnormalized posterior
unnormalized_posterior = prior * likelihoods

# Normalize the posterior
posterior = unnormalized_posterior / np.sum(unnormalized_posterior)

# Plot the prior, likelihood, and posterior distributions
plt.figure(figsize=(10, 6))
plt.plot(theta_values, prior, label='Prior', color='blue')
plt.plot(theta_values, likelihoods, label='Likelihood', color='red')
plt.plot(theta_values, posterior, label='Posterior', color='green')
plt.title('Generative Bayesian Model with Beta Prior')
plt.xlabel('Theta')
plt.ylabel('Density')
plt.legend()
plt.show()
```
