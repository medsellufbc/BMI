```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Generate synthetic data
np.random.seed(0)
data = np.random.exponential(scale=2, size=100)  # Exponential data with scale parameter 2

# Define the likelihood function (exponential distribution)
def likelihood(theta, data):
    return np.prod(theta * np.exp(-theta * data))

# Define the Beta prior for the parameter theta
alpha = 2  # Alpha parameter for the Beta distribution
beta_ = 2  # Beta parameter for the Beta distribution

# Perform Bayesian inference to estimate the posterior distribution
theta_values = np.linspace(0, 5, 1000)  # Range of theta values
prior = beta.pdf(theta_values, alpha, beta_)  # Prior distribution
likelihoods = [likelihood(theta, data) for theta in theta_values]  # Likelihood values
unnormalized_posterior = prior * likelihoods  # Unnormalized posterior
posterior = unnormalized_posterior / np.sum(unnormalized_posterior)  # Normalize the posterior

# Plot the prior, likelihood, and posterior distributions
plt.figure(figsize=(10, 6))
plt.plot(theta_values, prior, label='Prior', color='blue')
plt.plot(theta_values, likelihoods, label='Likelihood', color='red')
plt.plot(theta_values, posterior, label='Posterior', color='green')
plt.title('Exponential Bayesian Model with Beta Prior')
plt.xlabel('Theta')
plt.ylabel('Density')
plt.legend()
plt.show()
```
